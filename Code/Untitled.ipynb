```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
pm_b=[1,2,3,1,2,2,0,2,1,1]
cm_b=[2,3,2,4,1,3,0,3,1,2]
pm_b=[i+3 for i in pm_b]
cm_b=[i+3 for i in cm_b]
df=pd.DataFrame()
df['Phylogenetc to Mutational']=pm_b
df['Clonal to Mutational']=cm_b
sns.set()
fig=plt.figure(figsize=(10,7))
ax=sns.boxplot(data=df)
ax.set_title("MLTED Distance")
plt.show()
```
