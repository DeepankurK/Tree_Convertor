```python
import seaborn as sns
import matplotlib.pyplot as plt
ax = sns.boxplot(data=df.replace(0,np.nan)[[0,2,4,6,8,10]])
ax=sns.swarmplot(data=df.replace(0,np.nan)[[0,2,4,6,8,10]],color=".25")
ax.set(xlabel = 'Different Clustering Algorihms\n[0]:Affinity Propagation Clustering,[2]:Agglomerative Clustering, [4]:Birch Clustering,\n [6]:KMeans Clustering ,[8]:KMedoids Clustering and [10]:Spectral Clustering', ylabel = 'Value of Normalized Pairwise Distance', )
```

```python
df.replace(0,np.nan)[[1,3,5,7,9,11]].boxplot(figsize=(20,20))
# Column 0,3,6,9,12,15 are normalized pairwise distance for 6 algorithms used.
# Column 1,4,7,10,13,16 are V-Measure for 6 algorithms used.
# Column 2,5,8,11,14,17 are Fowlkes Mallows Score for 6 algorithms used.
# The six algorithms are in a sequential order (as used):-
#1. Affinity Propagation
#2. Agglomerative Clustering
#3. Birch
#4. KMeans
#5. KMedoids
#6. SpectralClustering
```
